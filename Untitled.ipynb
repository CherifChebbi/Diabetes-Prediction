**Project Description:**


This project is a machine learning application designed to predict whether a person is diabetic or not 
based on their health data. It uses the PIMA Indian Diabetes Dataset and applies a Support Vector Machine (SVM) 
with a linear kernel for classification. The system includes data preprocessing, model training, evaluation, 
and a predictive system for real-time input.

### Import Dependencies


```python
### Import Dependencies
# Importing necessary libraries for data manipulation, preprocessing, and modeling
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
import warnings



# Suppress warnings related to feature names mismatch
warnings.filterwarnings("ignore", message="X does not have valid feature names")
```

### Step 1: Load the Dataset

```python
# Load the diabetes dataset from a CSV file
df=pd.read_csv('diabetes.csv')
```

```python
# Display the first few rows of the dataset
print("First 5 rows of the dataset:")
print(df.head())
```

```python
# Display dataset shape
print("\nShape of the dataset:", df.shape)
```

```python
# Display dataset information
print("\nDataset information:")
print(df.info())
```

```python
# Analyse des valeurs nulles
print("\nNombre de valeurs manquantes dans chaque colonne :")
print(df.isnull().sum())
```

```python
# Display basic statistical summary
print("\nDataset statistics:")
print(df.describe())
```

```python
# Visualisation des distributions des variables
df.hist(bins=15, figsize=(15, 10), color='blue', edgecolor='black')
plt.suptitle("Distributions des variables", fontsize=16)
plt.show()
```

```python
#  Matrice de corrélation
correlation_matrix = df.corr()

plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Matrice de corrélation", fontsize=16)
plt.show()
```

```python
# Count the values in the 'Outcome' column (target variable)
print("\nDistribution of Outcome values:")
print(df['Outcome'].value_counts())
```

```python
# Group the dataset by 'Outcome' and calculate the mean for each group
print("\nMean values grouped by Outcome:")
print(df.groupby('Outcome').mean())
```

### Step 2: Data Preparation


```python
# Separate the features (X) and target (Y)
X = df.drop(columns='Outcome', axis=1)
Y = df['Outcome']
```

```python
# Display the features and target
print("\nFeatures (X):")
print(X.head())
print("\nTarget (Y):")
print(Y.head())

```

### Step 3: Data Standardization


```python
scaler = StandardScaler()
```

```python
# Standardize the feature data for better model performance
scaler = StandardScaler()
scaler.fit(X)  # Fit the scaler on the dataset
standardized_data = scaler.transform(X)
```

```python
# Update X with standardized data
X = standardized_data
```

```python
# Display the standardized feature data
print("\nStandardized Features (X):")
print(X[:5])
```

```python
X = standardized_data
Y = df['Outcome']
```

### Step 4: Train-Test Split


```python
# Split the dataset into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, test_size=0.2, stratify=Y, random_state=2
)
```

```python
# Display the shape of the training and testing sets
print("\nShape of the dataset after train-test split:")
print("Full dataset:", X.shape)
print("Training set:", X_train.shape)
print("Testing set:", X_test.shape)
```

### Step 5: Training the Model


```python
# Initialize the Support Vector Machine (SVM) Classifier with a linear kernel
classifier = svm.SVC(kernel='linear')
```

```python
# Train the classifier on the training data
classifier.fit(X_train, Y_train)
print("\nModel training complete.")
```

### Step 6: Model Evaluation


```python
# Evaluate the model on training data
X_train_prediction = classifier.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
print("\nAccuracy score on training data:", training_data_accuracy)
```

```python
# Evaluate the model on testing data
X_test_prediction = classifier.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
print("Accuracy score on testing data:", test_data_accuracy)
```

### Step 7: Predictive System


This section allows the user to input their health data and get a prediction 
on whether they are diabetic or not. The input data is processed and standardized 
before being passed to the trained SVM classifier for prediction.

```python
# Input data for prediction
input_data = (4, 110, 92, 0, 0, 37.6, 0.191, 30)

# Convert the input data to a numpy array
input_data_as_numpy_array = np.asarray(input_data)

# Reshape the array as we are predicting for one instance
input_data_reshaped = input_data_as_numpy_array.reshape(1, -1)

# Standardize the input data using the same scaler
std_data = scaler.transform(input_data_reshaped)

# Perform the prediction
prediction = classifier.predict(std_data)

# Display the prediction result
print("\nPrediction:")
if prediction[0] == 0:
    print("The person is not diabetic.")
else:
    print("The person is diabetic.")
```
